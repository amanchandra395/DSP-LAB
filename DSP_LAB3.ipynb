```python
import numpy as np
import matplotlib.pyplot as plt

def autocorr_signal(signal):
    autocorr = np.correlate(signal, signal, mode='full')
    return autocorr[len(autocorr)//2:]

def crosscorr_signals(signal1, signal2):
    crosscorr = np.correlate(signal1, signal2, mode='full')
    return crosscorr

# Get user input for signal 1
signal1 = list(map(float, input("Enter signal 1 : ").split()))

# Get user input for signal 2
signal2 = list(map(float, input("Enter signal 2 : ").split()))

autocorr1 = autocorr_signal(signal1)
autocorr2 = autocorr_signal(signal2)

crosscorr = crosscorr_signals(signal1, signal2)
    # Plotting the signals
plt.figure(figsize=(10,5))

plt.subplot(131)
plt.plot(signal1)
plt.title("Signal 1")
print("Signal 1")
print(signal1)

plt.subplot(132)
plt.plot(signal2)
plt.title("Signal 2")
print("Signal 2")
print(signal2)

plt.show()

i = int(input("Enter 1 for Cross-Correlation or 2 for Auto Correlation : "))
if(i==1):
    plt.subplot(133)
    plt.plot(crosscorr)
    plt.title("Cross-Correlation")
    print("Cross-Correlation")
    print(crosscorr)
    

elif(i==2):
    plt.show()
    plt.subplot(121)
    plt.plot(autocorr1)
    plt.title("Auto-Correlation of Signal 1")
    print("Auto-Correlation of Signal 1")
    print(autocorr1)
    
    plt.subplot(122)
    plt.plot(autocorr2)
    plt.title("Auto-Correlation of Signal 2")
    print("Auto-Correlation of Signal 2")
    print(autocorr2)
```
