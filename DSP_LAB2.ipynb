```python
import numpy as np
import matplotlib.pyplot as plt

def convolution(signal1, signal2):
    result = np.zeros(len(signal1) + len(signal2) - 1)
    for i in range(len(signal1)):
        for j in range(len(signal2)):
            result[i + j] += signal1[i] * signal2[j]
    return result

# Get user input for signal 1
s1 = list(map(float, input("Enter signal 1 : ").split()))

# Get user input for signal 2
s2 = list(map(float, input("Enter signal 2 : ").split()))

# Use numpy's convolve function to compute the convolution
conv_result = np.convolve(s1, s2)

# Custom implementation of convolution
conv_result_custom = convolution(s1, s2)

# Print the convolution results
print("Convolution using numpy's convolve function :")
print(conv_result)

print("Convolution using custom function :")
print(conv_result_custom)

plt.figure()
plt.plot(s1, label='S1')
plt.plot(s2, label='S2')
#plt.plot(conv_result, label='np.convolve Result')
plt.plot(conv_result_custom, label='R')
plt.legend()
plt.show()
```
